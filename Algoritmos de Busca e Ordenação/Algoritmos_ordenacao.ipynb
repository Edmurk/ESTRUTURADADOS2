## 1. Bubble Sort

O Bubble Sort é um algoritmo de ordenação simples que funciona repetidamente percorrendo a lista a ser ordenada, comparando elementos adjacentes e trocando-os de lugar se estiverem na ordem errada. A intuição por trás do Bubble Sort é semelhante ao processo de "borbulhamento" de elementos maiores para o topo (ou final) da lista, daí o nome "Bubble Sort".

### Intuição do Bubble Sort:

- Comparação de Pares Adjacentes:

  - Em cada iteração, o algoritmo compara pares de elementos adjacentes na lista.
  - Se o par está na ordem errada (por exemplo, se o primeiro elemento é maior que o segundo para uma ordenação crescente), os elementos são trocados de lugar.

- Borbulhamento:

  - Após cada iteração completa através da lista, o maior elemento "borbulha" para a sua posição correta no final da lista.
  - Isso acontece porque, a cada passagem, o maior elemento não trocado será movido para a sua posição final, assim como as bolhas de ar que sobem para o topo da água.

- Iterações Sucessivas:

  - O processo de comparação e troca é repetido para cada elemento da lista, excluindo o último elemento já ordenado a cada nova passagem.
  - Após a primeira passagem, o maior elemento está na posição correta, após a segunda passagem, os dois maiores elementos estão em suas posições corretas, e assim por diante.

- Convergência:

  - O algoritmo termina quando nenhuma troca é feita durante uma passagem completa, indicando que a lista está completamente ordenada.

## Exemplo Visual:

Considere a lista [5, 3, 8, 4, 2] a ser ordenada em ordem crescente.

  - Iteração 1:
    - Compara 5 e 3: Troca -> [3, 5, 8, 4, 2]
    - Compara 5 e 8: Não troca -> [3, 5, 8, 4, 2]
    - Compara 8 e 4: Troca -> [3, 5, 4, 8, 2]
    - Compara 8 e 2: Troca -> [3, 5, 4, 2, 8]
    - Agora, o maior elemento (8) está na posição correta.

  - Iteração 2:
    - Compara 3 e 5: Não troca -> [3, 5, 4, 2, 8]
    - Compara 5 e 4: Troca -> [3, 4, 5, 2, 8]
    - Compara 5 e 2: Troca -> [3, 4, 2, 5, 8]
    - Agora, os dois maiores elementos (5 e 8) estão nas posições corretas.

  - Iteração 3:
    - Compara 3 e 4: Não troca -> [3, 4, 2, 5, 8]
    - Compara 4 e 2: Troca -> [3, 2, 4, 5, 8]
    - Agora, os três maiores elementos (4, 5 e 8) estão nas posições corretas.

  - Iteração 4:
    - Compara 3 e 2: Troca -> [2, 3, 4, 5, 8]
    - Agora, a lista está completamente ordenada.

## Complexidade:
  - Melhor caso: O(n) - ocorre quando a lista já está ordenada.
  - Pior caso e caso médio: O(n²) - ocorre quando a lista está em ordem inversa.
  
## Vantagens e Desvantagens:
  - Vantagens:
    - Simplicidade de implementação.
    - Facilidade de entendimento.
    - Estável (não altera a ordem relativa de elementos iguais).
  - Desvantagens:

    - Ineficiente para listas grandes devido à complexidade O(n²).

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
            yield arr

def update_plot(arr, rects, iteration):
    for rect, val in zip(rects, arr):
        rect.set_height(val)
    plt.title(f"Bubble Sort - Iteration {iteration}")

def main():
    N = 20
    arr = np.random.randint(1, 20, N)
    generator = bubble_sort(arr)

    fig, ax = plt.subplots()
    ax.set_title("Bubble Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    iteration = [0]
    def animate(arr):
        update_plot(arr, rects, iteration[0])
        iteration[0] += 1

    anim = animation.FuncAnimation(
        fig, animate,
        frames=generator, repeat=False,
        interval=50, save_count=N * N
    )

    # Salvar a animação como um arquivo MP4
    anim.save('bubble_sort_animation.mp4', writer='ffmpeg', fps=30)

    # Salvar a animação como um arquivo GIF
    # anim.save('bubble_sort_animation.gif', writer='imagemagick', fps=30)

    plt.show()

if __name__ == "__main__":
    main()
```

# 2. Insertion Sort
## Algoritmo Insertion Sort
Insertion Sort é um algoritmo de ordenação simples e intuitivo, que constrói a lista ordenada um elemento de cada vez. Ele é eficiente para listas pequenas ou quase ordenadas, mas é ineficiente para listas grandes.

## Intuição:
### Construção Gradual:

- Começa com uma lista vazia (ou um elemento) e adiciona elementos da lista original um a um.
- Para cada novo elemento, ele é comparado com os elementos na lista ordenada e inserido na posição correta.

### Comparação e Deslocamento:

- Cada elemento é comparado com os elementos anteriores até encontrar sua posição correta.
- Os elementos maiores são deslocados uma posição para a direita para abrir espaço para o novo elemento.


## Complexidade
### Tempo:

- Melhor Caso: O(n)
  - Ocorre quando a lista já está ordenada. Em cada iteração, o algoritmo só precisa comparar o elemento atual com o anterior, sem deslocamentos.

- Pior Caso: O(n²)
  - Ocorre quando a lista está ordenada em ordem inversa. Cada novo elemento precisa ser comparado e deslocado por todos os elementos já ordenados.

- Caso Médio: O(n²)
  - No caso médio, a inserção de cada elemento precisa deslocar aproximadamente metade dos elementos já ordenados.
  
### Espaço:

- O(1)

  - Insertion Sort é um algoritmo in-place, o que significa que não requer espaço adicional proporcional ao tamanho da entrada.

### Vantagens:
- Simplicidade:

  - Fácil de entender e implementar.
  - Útil para ensinar conceitos básicos de algoritmos de ordenação.
  Eficiência para Pequenos Conjuntos de Dados:

  - É muito eficiente para listas pequenas.
  - Em listas pequenas, pode ser mais rápido que algoritmos mais complexos, como Merge Sort ou Quick Sort.

- Estabilidade:

  - É um algoritmo estável, o que significa que preserva a ordem relativa de elementos iguais.

  - Desempenho em Listas Quase Ordenadas:

  - Tem um desempenho muito bom quando aplicado a listas que já estão quase ordenadas.
  - Neste caso, o tempo de execução pode ser quase linear, O(n).

- Ordenação Online:

  - Pode ordenar uma lista à medida que os elementos são recebidos, útil em certos tipos de aplicações onde os dados chegam em tempo real.

### Desvantagens:
- Ineficiente para Grandes Conjuntos de Dados:

  - O tempo de execução O(n²) torna-o impraticável para listas grandes, especialmente quando comparado com algoritmos como Quick Sort ou Merge Sort, que têm complexidade O(n log n).

- Desempenho Consistente em Casos Médios:

  - Não é adequado para listas grandes e desordenadas, onde a complexidade quadrática leva a tempos de execução elevados.



## Exemplo Visual:
- Considere a lista [5, 2, 4, 6, 1, 3] a ser ordenada.

- Passo a Passo:
  - Começa com o segundo elemento (2), comparando e inserindo na posição correta:

  - Lista: [5, 2, 4, 6, 1, 3]
    - Ordenada: [2, 5, 4, 6, 1, 3]
    - Pega o próximo elemento (4) e insere na posição correta:

  - Lista: [2, 5, 4, 6, 1, 3]
    - Ordenada: [2, 4, 5, 6, 1, 3]
    - Continua o processo até que toda a lista esteja ordenada:

  - Lista: [2, 4, 5, 6, 1, 3]
    - Ordenada: [1, 2, 3, 4, 5, 6]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def insertion_sort(arr):
    frames = []
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
            frames.append(arr.copy())
        arr[j + 1] = key
        frames.append(arr.copy())
    return frames

def main():
    N = 50
    arr = np.random.randint(1, 100, N)
    frames = insertion_sort(arr.copy())

    fig, ax = plt.subplots()
    ax.set_title("Insertion Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(arr):
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot,
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/insertion_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 3. Selection Sort

## Algoritmo Selection Sort:
O Selection Sort é um algoritmo de ordenação simples que funciona encontrando repetidamente o menor (ou maior, dependendo da ordem) elemento da parte não ordenada da lista e movendo-o para a posição correta na parte ordenada da lista.

## Intuição:
- Divisão da Lista:

  - A lista é dividida em duas partes: a parte ordenada e a parte não ordenada.
  - Inicialmente, a parte ordenada é vazia e a parte não ordenada contém todos os elementos.

## Seleção do Menor Elemento:

  - Encontra-se o menor elemento da parte não ordenada.
  - Esse elemento é então trocado com o primeiro elemento da parte não ordenada, movendo-o para a última posição da parte ordenada.

## Expansão da Parte Ordenada:

- A fronteira entre as partes ordenada e não ordenada é movida um elemento à direita, expandindo a parte ordenada e reduzindo a parte não ordenada.

## Repetição:

- O processo é repetido até que toda a lista esteja ordenada.
Exemplo Visual:
- Considere a lista [5, 3, 8, 4, 2] a ser ordenada em ordem crescente.

  - Iteração 1:
    - A lista completa é a parte não ordenada: [5, 3, 8, 4, 2]
    - Menor elemento na parte não ordenada: 2
    - Troca 2 com o primeiro elemento: [2, 3, 8, 4, 5]
  - Iteração 2:
    - A parte ordenada agora é [2] e a parte não ordenada é [3, 8, 4, 5]
    - Menor elemento na parte não ordenada: 3
    - Troca 3 com o primeiro elemento da parte não ordenada: [2, 3, 8, 4, 5]
  - Iteração 3:
    - A parte ordenada agora é [2, 3] e a parte não ordenada é [8, 4, 5]
    - Menor elemento na parte não ordenada: 4
    - Troca 4 com o primeiro elemento da parte não ordenada: [2, 3, 4, 8, 5]
  - Iteração 4:
    - A parte ordenada agora é [2, 3, 4] e a parte não ordenada é [8, 5]
    - Menor elemento na parte não ordenada: 5
    - Troca 5 com o primeiro elemento da parte não ordenada: [2, 3, 4, 5, 8]
  - Iteração 5:
    - A parte ordenada agora é [2, 3, 4, 5] e a parte não ordenada é [8]
    - Menor (e único) elemento na parte não ordenada: 8
    - A lista está completamente ordenada.

## Complexidade:
### Tempo:

- Melhor Caso: O(n²)
- Pior Caso: O(n²)
- Caso Médio: O(n²)

- A complexidade de tempo é O(n²) para todos os casos porque o algoritmo sempre percorre a lista inteira para encontrar o próximo menor (ou maior) elemento, independentemente da ordem inicial dos elementos.

### Espaço:

- O(1)
- Selection Sort é um algoritmo in-place, o que significa que não requer espaço adicional proporcional ao tamanho da entrada, exceto por algumas variáveis temporárias.

### Vantagens:
- Simplicidade:
  - Fácil de entender e implementar.
  - Útil para ensinar conceitos básicos de algoritmos de ordenação.
- Desempenho Consistente:
  - O desempenho é previsível e não depende da ordem inicial dos elementos.
  - Sempre realiza O(n²) comparações e O(n) trocas, independentemente da entrada.
- Poucas Trocas:
  - Realiza o número mínimo possível de trocas (O(n)) entre elementos, o que pode ser vantajoso quando as trocas são relativamente caras em termos de tempo.

### Desvantagens:
- Ineficiente para Grandes Conjuntos de Dados:

- O tempo de execução O(n²) torna-o impraticável para listas grandes, especialmente quando comparado com algoritmos como Quick Sort ou Merge Sort, que têm complexidade O(n log n).
Não é Estável:

- Não preserva a ordem relativa dos elementos iguais, o que pode ser um problema em certas aplicações.
Desempenho Geral:

- Geralmente mais lento que outros algoritmos quadráticos como Insertion Sort, especialmente em listas parcialmente ordenadas.

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        yield arr

def update_plot(arr, rects, iteration):
    for rect, val in zip(rects, arr):
        rect.set_height(val)
    plt.title(f"Selection Sort - Iteration {iteration}")
    iteration[0] += 1

def main():
    N = 50
    arr = np.random.randint(1, 20, N)
    generator = selection_sort(arr)

    fig, ax = plt.subplots()
    ax.set_title("Selection Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    iteration = [0]
    anim = animation.FuncAnimation(
        fig, update_plot,
        fargs=(rects, iteration), frames=generator, repeat=False,
        interval=50, save_count=N * N
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/selection_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 4. Merge Sort
## Algoritmo Merge Sort
O Merge Sort é um algoritmo de ordenação baseado na técnica de divisão e conquista. A intuição por trás do Merge Sort é dividir a lista em partes menores até que cada parte tenha apenas um elemento, e então combinar (mesclar) essas partes de maneira ordenada para formar uma lista completamente ordenada.

## Intuição:

### Divisão:

- A lista é repetidamente dividida ao meio até que cada sublista tenha apenas um elemento.
- Sublistas de um único elemento são, por definição, ordenadas.

### Conquista:

- As sublistas são então mescladas (conquistadas) em pares, formando sublistas ordenadas.
- O processo de mesclagem envolve comparar os elementos de duas sublistas e combiná-los em uma nova sublista ordenada.

### Combinação:

- As sublistas mescladas são combinadas recursivamente até que toda a lista seja recombinada em uma única lista ordenada.

## Complexidade:
### Tempo:

  - Melhor Caso: $O(n \log n)$
  - Pior Caso: $O(n \log n)$
  - Caso Médio: $O(n \log n)$

  - A complexidade de tempo é $O(n \log n)$ para todos os casos porque o algoritmo sempre divide a lista em duas metades, ordena cada metade recursivamente e, em seguida, mescla as duas metades ordenadas.

### Espaço:

  - $O(n)$

  - Merge Sort não é um algoritmo in-place, o que significa que requer espaço adicional proporcional ao tamanho da entrada para armazenar as sublistas temporárias durante a mesclagem.
  
### Vantagens:

- Consistência de Desempenho:

- A complexidade de tempo $O(n \log n)$ é garantida para todos os casos (melhor, pior e médio), tornando-o eficiente e previsível.
Estabilidade:

- Merge Sort é um algoritmo estável, o que significa que preserva a ordem relativa dos elementos iguais.

- Divisão e Conquista:

  - O algoritmo utiliza a técnica de divisão e conquista, que pode ser eficiente para grandes conjuntos de dados e pode ser paralelizado.

  - Eficiente para Listas Grandes:

    - Merge Sort é muito eficiente para listas grandes e é a escolha preferida quando a estabilidade é importante e o espaço adicional está disponível.

### Desvantagens:

- Espaço Adicional:

  -Requer espaço adicional $O(n)$ para as sublistas temporárias, o que pode ser uma limitação em ambientes com memória restrita.
  Complexidade de Implementação:

  - Comparado a algoritmos simples como Insertion Sort ou Selection Sort, Merge Sort é mais complexo de implementar devido à necessidade de gerenciar as sublistas e a mesclagem.

- Desempenho em Listas Pequenas:

- Para listas pequenas, algoritmos como Insertion Sort podem ser mais eficientes devido ao menor overhead constante.

### Exemplo Visual:
- Considere a lista [38, 27, 43, 3, 9, 82, 10].

- Passo 1: Divisão
  - [38, 27, 43, 3, 9, 82, 10]
  - [38, 27, 43] e [3, 9, 82, 10]
  - [38], [27, 43] e [3, 9], [82, 10]
  - [38], [27], [43] e [3], [9], [82], [10]
- Passo 2: Conquista (Mesclagem)
  - [27, 38], [43] e [3, 9], [10, 82]
  - [27, 38, 43] e [3, 9, 10, 82]
- Passo 3: Combinação Final
  - [3, 9, 10, 27, 38, 43, 82]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def merge_sort(arr, left, right, frames):
    if left < right:
        middle = (left + right) // 2

        merge_sort(arr, left, middle, frames)
        merge_sort(arr, middle + 1, right, frames)

        merge(arr, left, middle, right, frames)

def merge(arr, left, middle, right, frames):
    n1 = middle - left + 1
    n2 = right - middle

    L = [0] * n1
    R = [0] * n2

    for i in range(0, n1):
        L[i] = arr[left + i]

    for j in range(0, n2):
        R[j] = arr[middle + 1 + j]

    i = 0
    j = 0
    k = left

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
        frames.append(arr.copy())

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
        frames.append(arr.copy())

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
        frames.append(arr.copy())

def main():
    N = 50
    arr = np.random.randint(1, 100, N)
    frames = []
    frames.append(arr.copy())

    merge_sort(arr, 0, N-1, frames)

    fig, ax = plt.subplots()
    ax.set_title("Merge Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(arr, rects):
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot, fargs=(rects,),
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/merge_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 5. Quick Sort
## Algoritmo Quick Sort
O Quick Sort é um algoritmo de ordenação eficiente que utiliza a técnica de divisão e conquista. A intuição por trás do Quick Sort é escolher um "pivô" e particionar a lista de forma que todos os elementos menores que o pivô fiquem à esquerda e todos os elementos maiores que o pivô fiquem à direita. O processo é então repetido recursivamente para as sublistas à esquerda e à direita do pivô.

## Intuição:

### Escolha do Pivô:

- Escolhe-se um elemento da lista como pivô. O pivô pode ser escolhido de várias maneiras (primeiro elemento, último elemento, elemento aleatório, etc.).

### Particionamento:

- Reorganiza-se a lista de forma que todos os elementos menores que o pivô fiquem à esquerda do pivô e todos os elementos maiores que o pivô fiquem à direita.

### Recursão:

- Aplica-se o Quick Sort recursivamente às sublistas à esquerda e à direita do pivô.

### Conquista:

- A ordenação é obtida quando todas as sublistas estão ordenadas.

## Complexidade:
### Tempo:

- Melhor Caso: $O(n \log n)$
  - Ocorre quando o pivô escolhido divide a lista de maneira balanceada.
- Pior Caso: $O(n_2)$
  - Ocorre quando o pivô escolhido é sempre o maior ou o menor elemento, dividindo a lista de forma muito desigual.
- Caso Médio: $O(n \log n)$
Em média, o Quick Sort realiza $O(n \log n)$ comparações, tornando-o eficiente para a maioria dos conjuntos de dados.

### Espaço:

- $O(n \log n)$
- O Quick Sort é um algoritmo in-place, o que significa que ele requer apenas espaço adicional para a pilha de chamadas recursivas. No melhor caso, a profundidade da pilha é $O(n \log n)$, mas pode ser
$O(n)$ no pior caso.

### Vantagens:
- Desempenho Médio Excelente:

- Em média, o Quick Sort é mais rápido que outros algoritmos de ordenação como Merge Sort e Heap Sort devido a seu menor overhead constante.

- In-Place:

- O Quick Sort é um algoritmo in-place, o que significa que ele não requer espaço adicional significativo além do necessário para a pilha de chamadas recursivas.

### Facilidade de Implementação:

- Relativamente simples de implementar e entender, especialmente com a escolha correta de pivô.

###Versatilidade:

- Funciona bem com grandes conjuntos de dados e pode ser facilmente adaptado para várias condições de entrada.
Desvantagens:
- Pior Caso $O(n_2)$: O pior caso de complexidade de tempo é
$O(n_2)$, o que pode ocorrer com uma escolha inadequada de pivô (como sempre escolher o maior ou o menor elemento).

- Não Estável:

  - O Quick Sort não é um algoritmo estável, o que significa que a ordem relativa dos elementos iguais pode não ser preservada.

- Requer Cuidados com a Escolha do Pivô:

  - A eficiência do algoritmo depende fortemente da escolha do pivô. Métodos como a "mediana de três" ou pivôs aleatórios são frequentemente usados para melhorar o desempenho.

### Exemplo Visual:
- Considere a lista [10, 7, 8, 9, 1, 5] a ser ordenada em ordem crescente.

- Passo 1: Escolha do Pivô
  - Escolhe-se o último elemento como pivô: 5.
- Passo 2: Particionamento
  - Reorganiza-se a lista: [1, 5, 8, 9, 10, 7].
- Passo 3: Recursão
  - Aplica-se o Quick Sort às sublistas [1] e [8, 9, 10, 7].
- Passo 4: Conquista
  - A lista é ordenada recursivamente até que todas as sublistas estejam ordenadas.

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def quick_sort(arr, low, high, frames):
    if low < high:
        pi = partition(arr, low, high, frames)
        quick_sort(arr, low, pi - 1, frames)
        quick_sort(arr, pi + 1, high, frames)

def partition(arr, low, high, frames):
    i = low - 1
    pivot = arr[high]

    for j in range(low, high):
        if arr[j] < pivot:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
            frames.append(arr.copy())

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    frames.append(arr.copy())
    return i + 1

def main():
    N = 50
    arr = np.random.randint(1, 100, N)
    frames = []
    frames.append(arr.copy())

    quick_sort(arr, 0, N-1, frames)

    fig, ax = plt.subplots()
    ax.set_title("Quick Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(arr, rects):
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot, fargs=(rects,),
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/quick_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 6. Heap Sort

## Algoritmo Heap Sort
O Heap Sort é um algoritmo de ordenação baseado na estrutura de dados chamada heap, que é uma árvore binária completa. A intuição por trás do Heap Sort é transformar a lista em um max-heap (ou min-heap) e, em seguida, extrair o maior (ou menor) elemento repetidamente para construir a lista ordenada.

## Intuição:
### Construção do Heap:

- Constrói-se um max-heap a partir da lista original. Em um max-heap, cada pai é maior ou igual a seus filhos.

### Extração e Reajuste:

- O maior elemento (a raiz do heap) é trocado com o último elemento da lista. A lista é então "reduzida" e o heap é ajustado para manter a
propriedade de max-heap.
- O processo é repetido até que toda a lista esteja ordenada.

### Exemplo Visual:
- Considere a lista [4, 10, 3, 5, 1].

- Passo 1: Construção do Max-Heap
  - Lista inicial: [4, 10, 3, 5, 1]
  - Max-Heap: [10, 5, 3, 4, 1]
- Passo 2: Extração e Reajuste
  - Troca 10 com 1: [1, 5, 3, 4, 10]
    - Reajusta o heap: [5, 4, 3, 1, 10]
  - Troca 5 com 1: [1, 4, 3, 5, 10]
    - Reajusta o heap: [4, 1, 3, 5, 10]
  - Troca 4 com 1: [1, 3, 4, 5, 10]
    - Reajusta o heap: [3, 1, 4, 5, 10]
  - Troca 3 com 1: [1, 3, 4, 5, 10]
  - Lista ordenada: [1, 3, 4, 5, 10]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def heapify(arr, n, i, frames):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        frames.append(arr.copy())
        heapify(arr, n, largest, frames)

def heap_sort(arr, frames):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i, frames)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        frames.append(arr.copy())
        heapify(arr, i, 0, frames)

def main():
    N = 50
    arr = np.random.randint(1, 100, N)
    frames = []
    frames.append(arr.copy())

    heap_sort(arr, frames)

    fig, ax = plt.subplots()
    ax.set_title("Heap Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(arr, rects):
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot, fargs=(rects,),
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/heap_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 7. Radix Sort

## Algoritmo Radix Sort

O Radix Sort é um algoritmo de ordenação não comparativo que ordena os números processando-os dígito por dígito, começando do dígito menos significativo para o mais significativo (LSD - Least Significant Digit) ou vice-versa (MSD - Most Significant Digit). É eficiente para ordenar números inteiros e outras chaves que podem ser representadas como uma sequência de dígitos.

## Intuição:
### Ordenação por Dígitos:

- Em cada iteração, os números são agrupados com base no valor de um dígito específico.
- Para LSD Radix Sort, a ordenação começa pelo dígito menos significativo.

### Estabilidade:

- O algoritmo utiliza uma ordenação estável (como Counting Sort) para cada dígito, garantindo que a ordem relativa dos números com o mesmo valor de dígito seja preservada.

### Combinação dos Resultados:

- Após processar todos os dígitos, os números estarão ordenados.

## Exemplo Visual:
- Considere a lista [170, 45, 75, 90, 802, 24, 2, 66] para ordenar em ordem crescente.

- Passo 1: Ordenação pelo dígito menos significativo (unidades)
  - Lista após a ordenação: [170, 90, 802, 2, 24, 45, 75, 66]
- Passo 2: Ordenação pelo segundo dígito (dezenas)
  - Lista após a ordenação: [802, 2, 24, 45, 66, 170, 75, 90]
- Passo 3: Ordenação pelo terceiro dígito (centenas)
  - Lista após a ordenação: [2, 24, 45, 66, 75, 90, 170, 802]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def counting_sort(arr, exp, frames):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    for i in range(n):
        index = arr[i] // exp
        count[index % 10] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = arr[i] // exp
        output[count[index % 10] - 1] = arr[i]
        count[index % 10] -= 1
        i -= 1

    for i in range(n):
        arr[i] = output[i]
        frames.append(arr.copy())

def radix_sort(arr, frames):
    max1 = max(arr)

    exp = 1
    while max1 // exp > 0:
        counting_sort(arr, exp, frames)
        exp *= 10

def main():
    N = 50
    arr = np.random.randint(1, 1000, N)
    frames = []
    frames.append(arr.copy())

    radix_sort(arr, frames)

    fig, ax = plt.subplots()
    ax.set_title("Radix Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(arr, rects):
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot, fargs=(rects,),
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/radix_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 8. Counting Sort
## Algoritmo Counting Sort
Counting Sort é um algoritmo de ordenação não comparativo, eficiente para ordenar números inteiros dentro de um intervalo específico. A intuição por trás do Counting Sort é contar a ocorrência de cada elemento da lista e usar essa contagem para determinar a posição correta de cada elemento na lista ordenada.

## Intuição:
### Contagem das Ocorrências:

- Conta o número de vezes que cada valor aparece na lista e armazena essas contagens em um array auxiliar (contagem).

### Acumulação das Contagens:

- Transforma o array de contagens em um array acumulado para determinar a posição final de cada elemento.

### Construção da Lista Ordenada:

- Usa o array acumulado para colocar cada elemento na sua posição final correta na lista ordenada, garantindo a estabilidade (a ordem relativa dos elementos iguais é preservada).

## Exemplo Visual:
- Considere a lista [4, 2, 2, 8, 3, 3, 1].

- Passo 1: Contagem das Ocorrências
  - Lista original: [4, 2, 2, 8, 3, 3, 1]
  - Contagem: [0, 1, 2, 2, 1, 0, 0, 0, 1] (índice representa o valor)
- Passo 2: Acumulação das Contagens
  - Contagem acumulada: [0, 1, 3, 5, 6, 6, 6, 6, 7]
  - Passo 3: Construção da Lista Ordenada
- Lista ordenada: [1, 2, 2, 3, 3, 4, 8]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def counting_sort(arr, frames):
    max_val = max(arr)
    count = [0] * (max_val + 1)
    output = [0] * len(arr)

    # Contagem das ocorrências
    for num in arr:
        count[num] += 1
        frames.append((arr.copy(), count.copy(), output.copy()))

    # Acumulação das contagens
    for i in range(1, len(count)):
        count[i] += count[i - 1]
        frames.append((arr.copy(), count.copy(), output.copy()))

    # Construção da lista ordenada
    for num in reversed(arr):
        output[count[num] - 1] = num
        count[num] -= 1
        frames.append((arr.copy(), count.copy(), output.copy()))

    for i in range(len(arr)):
        arr[i] = output[i]
        frames.append((arr.copy(), count.copy(), output.copy()))

def main():
    N = 50
    arr = np.random.randint(1, 100, N)
    frames = []
    frames.append((arr.copy(), [0] * 100, [0] * N))

    counting_sort(arr, frames)

    fig, ax = plt.subplots()
    ax.set_title("Counting Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")
    count_rects = ax.bar(range(100), [0] * 100, align="edge", color='r', alpha=0.5)
    output_rects = ax.bar(range(len(arr)), [0] * len(arr), align="edge", color='g', alpha=0.5)

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(frame_data):
        arr, count, output = frame_data
        for rect, val in zip(rects, arr):
            rect.set_height(val)
        for rect, val in zip(count_rects, count):
            rect.set_height(val)
        for rect, val in zip(output_rects, output):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot,
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/counting_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 9. Bucket Sort
## Algoritmo Bucket Sort
O Bucket Sort é um algoritmo de ordenação que distribui os elementos de uma lista em vários "baldes" (ou buckets) e então ordena cada balde individualmente. Depois, os baldes são concatenados para formar a lista ordenada final. Este algoritmo é especialmente eficiente para dados que são distribuídos uniformemente.

## Intuição:
### Distribuição em Baldes:

- Os elementos da lista são distribuídos em vários baldes com base em algum critério (por exemplo, valor).
- Cada balde contém elementos que pertencem a um intervalo específico de valores.

### Ordenação dos Baldes:

- Cada balde é ordenado individualmente usando outro algoritmo de ordenação (geralmente Insertion Sort ou Quick Sort).

### Concatenar Baldes:

- Os baldes ordenados são concatenados para formar a lista ordenada final.

## Exemplo Visual:
- Considere a lista [0.78, 0.17, 0.39, 0.26, 0.72, 0.94, 0.21, 0.12, 0.23, 0.68] a ser ordenada.

- Passo 1: Distribuição em Baldes
  - Distribuir os elementos em baldes baseados em seus valores:
    - Balde 1: [0.12, 0.17, 0.21, 0.23, 0.26]
    - Balde 2: [0.39, 0.68, 0.72, 0.78]
    - Balde 3: [0.94]
- Passo 2: Ordenação dos Baldes
  - Ordenar cada balde individualmente:
    - Balde 1: [0.12, 0.17, 0.21, 0.23, 0.26]
    - Balde 2: [0.39, 0.68, 0.72, 0.78]
    - Balde 3: [0.94]
- Passo 3: Concatenar Baldes
  - Concatenar os baldes ordenados para formar a lista final ordenada:
    - Lista ordenada: [0.12, 0.17, 0.21, 0.23, 0.26, 0.39, 0.68, 0.72, 0.78, 0.94]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def insertion_sort(bucket):
    for i in range(1, len(bucket)):
        up = bucket[i]
        j = i - 1
        while j >= 0 and bucket[j] > up:
            bucket[j + 1] = bucket[j]
            j -= 1
        bucket[j + 1] = up
    return bucket

def bucket_sort(arr, frames):
    bucket = []

    for i in range(len(arr)):
        bucket.append([])

    for j in arr:
        index_b = int(10 * j)  # Assumindo que os valores são de 0 a 1
        bucket[index_b].append(j)

    for i in range(len(arr)):
        bucket[i] = insertion_sort(bucket[i])
        frames.append((arr.copy(), bucket.copy()))

    k = 0
    for i in range(len(arr)):
        for j in range(len(bucket[i])):
            arr[k] = bucket[i][j]
            k += 1
            frames.append((arr.copy(), bucket.copy()))

def main():
    N = 50
    arr = np.random.rand(N)
    frames = []
    frames.append((arr.copy(), [[] for _ in range(N)]))

    bucket_sort(arr, frames)

    fig, ax = plt.subplots()
    ax.set_title("Bucket Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, 1)

    def update_plot(frame_data):
        arr, buckets = frame_data
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot,
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/bucket_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.show()

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```

# 10. Tim Sort
## Algoritmo Tim Sort
Tim Sort é um algoritmo de ordenação híbrido derivado de Merge Sort e Insertion Sort. Foi desenvolvido por Tim Peters em 2002 para a implementação do método de ordenação padrão da linguagem Python. Tim Sort é projetado para funcionar bem em muitos tipos de dados do mundo real, que frequentemente têm segmentos (runs) pré-ordenados.

## Intuição:
### Divisão em Runs:

- A lista é dividida em pequenos segmentos chamados "runs". Cada run é ordenado individualmente usando Insertion Sort.
- Runs são sequências contíguas de elementos já ordenados ou quase ordenados.

### Ordenação dos Runs:

- Cada run é ordenado usando Insertion Sort, que é eficiente para pequenos conjuntos de dados.

### Mesclagem dos Runs:

- Os runs ordenados são mesclados usando uma técnica semelhante ao Merge Sort para formar uma lista ordenada maior.
- A mesclagem é feita de maneira otimizada para minimizar o número de cópias e comparações.

## Exemplo Visual:
- Considere a lista [5, 21, 7, 23, 19, 8, 14, 29, 18].

- Passo 1: Divisão em Runs
  - Dividir a lista em runs:
    - Run 1: [5, 21] (já ordenado)
    - Run 2: [7, 23] (já ordenado)
    - Run 3: [19] (elemento único, já ordenado)
    - Run 4: [8, 14, 29] (já ordenado)
    - Run 5: [18] (elemento único, já ordenado)
    - Passo 2: Ordenação dos Runs
    - Runs individuais são pequenos, então já estão ordenados.
- Passo 3: Mesclagem dos Runs
  - Mesclar os runs ordenados para formar uma lista maior e ordenada:
    - Mesclagem 1: [5, 21] e [7, 23] -> [5, 7, 21, 23]
    - Mesclagem 2: [19] e [8, 14, 29] -> [8, 14, 19, 29]
    - Mesclagem 3: [8, 14, 19, 29] e [18] -> [8, 14, 18, 19, 29]
    - Mesclagem Final: [5, 7, 21, 23] e [8, 14, 18, 19, 29] -> [5, 7, 8, 14, 18, 19, 21, 23, 29]

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

def insertion_sort(arr, left, right):
    for i in range(left + 1, right + 1):
        key = arr[i]
        j = i - 1
        while j >= left and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def merge(arr, l, m, r):
    len1, len2 = m - l + 1, r - m
    left, right = [], []
    for i in range(0, len1):
        left.append(arr[l + i])
    for i in range(0, len2):
        right.append(arr[m + 1 + i])

    i, j, k = 0, 0, l
    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    while i < len1:
        arr[k] = left[i]
        k += 1
        i += 1

    while j < len2:
        arr[k] = right[j]
        k += 1
        j += 1

def tim_sort(arr, frames, run=32):
    n = len(arr)
    for start in range(0, n, run):
        end = min(start + run - 1, n - 1)
        insertion_sort(arr, start, end)
        frames.append(arr.copy())

    size = run
    while size < n:
        for left in range(0, n, 2 * size):
            mid = min(n - 1, left + size - 1)
            right = min((left + 2 * size - 1), (n - 1))
            if mid < right:
                merge(arr, left, mid, right)
                frames.append(arr.copy())
        size = 2 * size

def main():
    N = 50
    arr = np.random.randint(1, 100, N)
    frames = []
    frames.append(arr.copy())

    tim_sort(arr, frames)

    fig, ax = plt.subplots()
    ax.set_title("Tim Sort")
    rects = ax.bar(range(len(arr)), arr, align="edge")

    ax.set_xlim(0, N)
    ax.set_ylim(0, int(1.1 * max(arr)))

    def update_plot(arr, rects):
        for rect, val in zip(rects, arr):
            rect.set_height(val)

    anim = animation.FuncAnimation(
        fig, func=update_plot, fargs=(rects,),
        frames=frames, repeat=False, interval=100
    )

    # Salvar a animação como um arquivo MP4
    anim.save('/content/tim_sort_animation.mp4', writer='ffmpeg', fps=30)

    plt.close()  # Fechar a figura para evitar exibição dupla

if __name__ == "__main__":
    main()
```
